```python
new_df=movies[["movie_id","title","tags"]]
new_df.head()
```

```python
# convert tags column's list to string

new_df["tags"]=new_df["tags"].apply(lambda x: " ".join(x))
```

```python
# convert all strings of tags colum to lower case

new_df['tags']=new_df["tags"].apply(lambda x: x.lower())
```

```python
new_df.head()
```

```python
# in tags feature there are sames words but in different forms. so we need to make them same
# use stemming technique to get the result
# (love,loving,loves)=love,love,love)
```

```python
import nltk
from nltk.stem.porter import PorterStemmer
ps=PorterStemmer()
```

```python
def stem(text):
  y=[]
  for i in text.split():
    y.append(ps.stem(i))

  return " ".join(y)
```

```python
new_df["tags"]=new_df['tags'].apply(stem)
new_df.head()
```

# to get simmmilar names from tags we need to use text-vecorization technique
# before vectoriztion technique we need to remove stop words from considearation

```python
from sklearn.feature_extraction.text import CountVectorizer
cv=CountVectorizer(max_features=5000,stop_words="english")
```

```python
vectors=cv.fit_transform(new_df["tags"]).toarray()
```

```python
vectors
```

# we need to calculate the distance between to vectors. this is a high dimensional data,(5000),so we need to use cosine distance in instead of eucledian distance


```python
from sklearn.metrics.pairwise import cosine_similarity
```

```python
# calculate cosine distance
similarity=cosine_similarity(vectors)
```
